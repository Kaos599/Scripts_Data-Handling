```python
import torch
from peft import PeftModel
from transformers import AutoModelForCausalLM, AutoTokenizer
import transformers
adapters_name = r"C:\Users\harsh\Downloads\cascadejv1"
model_name = "Llama-2-7b-chat-hf"
model_path = r"C:\Users\harsh\Downloads\Llama-2-7b-chat-hf"



device = "cuda" # the device to load the model onto
```

```python
bnb_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    quantization_config=bnb_config,
    device_map='auto'
)
```

```python
model = PeftModel.from_pretrained(model, adapters_name)
#model = model.merge_and_unload()
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.bos_token_id = 1

stop_token_ids = [0]

print(f"Successfully loaded the model {model_name} into memory")
```

```python

text = "Create a study plan for 'Fundamentals of AI' covering Machine Learning, Neural Networks, Robotics, Natural Language Processing, and Computer Vision. The study period is from May 15th to June 8th, with availability on Monday, Tuesday, Wednesday, and Friday from 10 AM to 1 PM. User ability score: 10."

encoded = tokenizer(text, return_tensors="pt", add_special_tokens=False)
model_input = encoded
model.to(device)
generated_ids = model.generate(**model_input, max_new_tokens=200, do_sample=True)
decoded = tokenizer.batch_decode(generated_ids)
print(decoded[0])
```
